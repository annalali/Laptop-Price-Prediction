```python
# Convert OpSys column
# df2['OpSys'].unique()
os_dict1 = {
    "Windows 7": "win7",
    "Windows 10": "win10",#
    "macOS": "mac", #
    "No OS": "No OS", #
    "Mac OS X" :"Mac OS X", #
    "Linux": "Linux",#
    "Android" : "Android",#
    "Windows 10 S": "win10_sec",#
    "Chrome OS": "chrome"#
}
df2["OpSys"]= df2["OpSys"].apply(lambda x: os_dict1.get(x, np.nan))
#=================================================================
# Convert RAM colunm to number
# df2['Ram'].unique() 
#df2.Ram = pd.to_numeric(df2.Ram.str.replace('GB', ''),errors="coerce")
df2['Ram'] = df2['Ram'].apply(lambda x: x.split("GB")[0]).astype("int64")
#=================================================================
# Convert from EUR to USD
df2["Price-USD"] = df2["Price_euros"]*1.13
```

### Preprocessing the data
### Cleaning the data 

```python
df4.head(2)
```

```python
# Create two columns from Name column
df4["Company"]= df4['Name'].str.split(' ').str[0] #.replace('-' , ' ' )
df4['Product'] = df4.Name.str.split(' ').str[1:3].str.join(' ')
```

```python
df4['RAM'].unique()
```

```python
# Delete the word RAM from the RAM column
# df4.RAM = df4.RAM.str.split(' ').str[0:3].str.join(' ')
# Convert RAM colunm to number
df4.RAM = pd.to_numeric(df4.RAM.str.split(' ').str[0],errors="coerce")
## df4["RAM"] = df4["RAM"].apply(lambda x: x.split("GB")[0]).astype("int64")
## df4.RAM = df4.RAM.str.split(' ').str[0]#.astype("int64")
## df4.RAM = df4.RAM.astype("int64")
#==================================================
# Convert Operating System column
# df4['Operating System'].unique()
os_dict = {
    "64 bit Windows 10 Operating System": "64-bit win10",
    "Windows 10 Operating System": "win10",
    "Mac OS Operating System": "mac",
    "Pre-installed Genuine Windows 10 Operating System (Includes Built-in Security, Free Automated Updates, Latest Features)":
        "win10_sec",
    "DOS Operating System":"dos",
    "64 bit Chrome Operating System": "64-bit chrome"
}
df4["Operating System"]= df4["Operating System"].apply(lambda x: os_dict.get(x, np.nan))
#=======================================================
# Convert Display size to number
df4.Display = pd.to_numeric(
    df4.Display.str.split(' ').str[0].replace(r'[^0-9.]','',regex=True),
    errors="coerce")
# Fix the cm convert to inches
df4.loc[df4.Display == 15.6,"Display"] = 39.624 # 39.624 / 2.54
df4.Display = df4.Display / 2.54
# Fill NAs with mean
disp_med = df4.Display.mean()
df4.Display = df4.Display.fillna(value=disp_med)
#======================================================
# Convert price to a number
df4.Price = df4.Price.str.replace(r'[^0-9]','',regex=True).astype('float')
# Convert from INR to USD
df4["Price-USD"] = df4["Price"]*0.013
```
